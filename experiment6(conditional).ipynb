```python
# 6. Write code for verifying whether a year is a leap year.
y = int(input("Enter year to check for leap year: "))
if (y % 4 == 0 and y % 100 != 0) or (y % 400 == 0):
  print("Leap Year.")
else:
  print("Not a leap year.")
```
